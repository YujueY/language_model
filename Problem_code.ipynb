```python
# 切分语法
def generate_grammer(grammar_str, split = '='):
    grammer = {}
    for t in grammar_str.split('\n'):
        if not t:
            continue
        else:
            expr, cont = t.split(split)
            grammer[expr.strip()] = [s.split() for s in cont.split('|')]
    return grammer
```

```python
import random
# 生成一个句子
def generate(grammar_str, target):
    if target not in grammer:
        return target
#     non = generate(grammer, t) for t in random.choice(grammer[target])
    return ''.join(generate(grammer, t) for t in random.choice(grammer[target]))
```

```python
# 生成n个句子
def generate_n(grammar_str, target, n):
    i = 0
    sentences = []
    while i < n:
        i += 1
        sentence = generate(grammar_str, target)
        sentences.append(sentence)
    return sentences
```

```python
grammer = generate_grammer(host)
generate_n(grammer, 'host', n=20)
```

2. 使用新数据源完成语言模型的训练

```python
import re
# 读取数据并去除噪音
data = []
with open('C:/Users/asus/Desktop/lesson/first_lesson/train.txt', "r", encoding = 'UTF-8') as f:
    for i in f:
        i = re.findall('\w+', i)
        i = ''.join(i)
        data.append(i)
# print(len(data))
```

```python
import jieba
# 切分数据
words = []
for i, line in enumerate(data):
#     if i % 100 == 0: print(i)
    # replace 10000 with a big number when you do your homework. 
    words += [str(t) for t in jieba.cut(line)]
```

```python
from collections import Counter
# 词频统计
words_count = Counter(words)
print(words_count)
```

```python
# 生成2元词组
words_2_GRAM = [''.join(words[i:i+2]) for i in range(len(words[:-2]))]
```

```python
# 二元词组词频统计
words_count_2 = Counter(words_2_GRAM)
```

```python
# 计算二元词组概率
def prob_2(word1, word2):
    if word1 + word2 in words_count_2: 
        return words_count_2[word1+word2] / len(words_2_GRAM)
    else:
        return 1 / len(words_2_GRAM)
```

```python
# 计算句子概率
import jieba
def get_probablity(sentence):
    
    words = []
    
    words += [str(t) for t in jieba.cut(sentence)]
    
    sentence_pro = 1
    
    for i, word in enumerate(words[:-1]):
        next_ = words[i+1]
        
        probability = prob_2(word, next_)
        
        sentence_pro *= probability
    
    return sentence_pro
```

3.获得最优质的的语言

```python
# 生成最佳句子及概率
def generate_best(grammar_str, target, n):
    i = 0
    sens_pro = []
    while i < n:
        i += 1
        sentence = generate(grammar_str, target)
        pro = get_probablity(sentence)
        sen_pro = (sentence, pro)
        sens_pro.append(sen_pro)
    sen_best = sorted(sens_pro, key=lambda x: x[1], reverse=True)[0]
    return sen_best
```

```python
generate_best(host, 'host', 10)
```
